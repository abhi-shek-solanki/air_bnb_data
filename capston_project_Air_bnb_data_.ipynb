```python

combining_neighboorhoods = air_bnb_new.groupby(["neighbourhood_group","neighbourhood"])["room_type"].value_counts()
print(combining_neighboorhoods)
```

```python
# Assuming combining_neighboorhoods is a Series or DataFrame obtained from the groupby operation

# Convert the grouped data to a DataFrame for visualization
combining_neighboorhoods_df = combining_neighboorhoods.reset_index(name='count')

# Create a pivot table to prepare data for the heatmap
pivot_df = combining_neighboorhoods_df.pivot_table(index=['neighbourhood_group', 'neighbourhood'], columns='room_type', values='count', fill_value=0)

# Plotting the heatmap
plt.figure(figsize=(12, 14))
sns.heatmap(pivot_df, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Room Type Counts by Neighbourhood Group and Neighbourhood')
plt.xlabel('Room Type')
plt.ylabel('Neighbourhood Group, Neighbourhood')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 5 which host name has booked the rooms most and in which neighboorhood

```python
# now lets check the host name means compare it

name = air_bnb.loc[air_bnb["host_id"].isin([219517861,107434423,30283594,137358866,16098958,12243051,22541573,200380610,7503643,1475015,120762452,2856748,205031545,190921808,26377263])]
name = name.groupby(["host_name","neighbourhood_group","neighbourhood"])["room_type"].value_counts()
print(name)


import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'name' is a DataFrame obtained from the groupby operation

# Reset the index to turn the grouped data into a DataFrame
name_df = name.reset_index(name='count')

# Create a pivot table to prepare data for the heatmap
pivot_df = name_df.pivot_table(index=['host_name', 'neighbourhood_group', 'neighbourhood'], columns='room_type', values='count', fill_value=0)

# Plotting the heatmap
plt.figure(figsize=(12, 20))
sns.heatmap(pivot_df, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Room Type Counts by Host, Neighbourhood Group, and Neighbourhood')
plt.xlabel('Room Type')
plt.ylabel('Host, Neighbourhood Group, Neighbourhood')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
